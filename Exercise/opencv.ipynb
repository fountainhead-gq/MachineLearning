```python
# -*- coding: utf-8 -*-

import cv2

# imread默认读取的是RGB格式，所以即使原图像是灰度图，读出来仍然是三个通道
# img = cv2.imread(r'C:\Users\shucun\Pictures\Camera Roll\fx.png',cv2.IMREAD_GRAYSCALE) # 0 读入灰色图片
img = cv2.imread(r'C:\Users\shucun\Pictures\Camera Roll\fx.png',1) # 1: cv2.IMREAD_COLOR, -1:cv2.IMREAD_UNCHANGED

cv2.imshow('image',img) # 建立image窗口显示图片
k = cv2.waitKey(0) # 无限期等待输入
 
if k==27: # wait for ESC key to exit
    cv2.destroyAllWindows() # 删除建立的全部窗口  
elif k==ord('s'): # wait for 's' key to save and exit
    cv2.imwrite('test.png',img)
    cv2.destroyAllWindows()

print(img.shape) # (749, 1096, 3)

h, w = img.shape[:2]
print(h,w) # 打印图像尺寸 
```
